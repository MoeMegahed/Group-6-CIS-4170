```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
plt.style.use('seaborn')
```

```python
df = pd.read_csv("Unicorn_Companies.csv")
```

```python
df.head()
```

```python
df_top5_country = pd.DataFrame(df.Country.value_counts().head(5))
```

```python
df_top5_country
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x = df.Country.value_counts().head(5).index,
            y = df.Country.value_counts().head(5).values)
plt.xticks(rotation = 45)
plt.title("Top 5 Countries",fontweight="bold")
plt.show()
```

```python
df_top_industry = pd.DataFrame(df["Industry"].value_counts().head(10))
```

```python
df_top_industry
```

```python
plt.figure(figsize=(7,5))
sns.barplot(y = df["Industry"].value_counts().head(10).index,
            x = df["Industry"].value_counts().head(10).values, 
            orient = 'h')
plt.xticks()
plt.title("Top 10 Industries",fontweight="bold")
plt.show()
```

```python
df_top_cities = pd.DataFrame(df["City"].value_counts().head(10))
```

```python
df_top_cities
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x = df["City"].value_counts().head(10).index,
            y = df["City"].value_counts().head(10).values, )
plt.xticks(rotation = 45)
plt.title("Top 10 Cities",fontweight="bold")
plt.show()
```
